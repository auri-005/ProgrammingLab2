### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
#ESEMPIO1
#codice normale
n = [1,2,3,4,5]
quadrati = []
for x in n:
    quadrati.append(x**2)
quadrati

#list comprehension
quadrati = [x**2 for x in n]
quadrati
```

```python
#ESEMPIO2
#codice normale
nomi = ["ali", "sara", "luca"]
maiuscolo = []
for nome in nomi:
    maiuscolo.append(nome.upper())
maiuscolo

#list comprehension
maiuscolo = [nome.upper() for nome in nomi]
maiuscolo
```

```python
#ESEMPIO3
#codice normale
pari = []
for x in range(10):
    if x%2==0:
        pari.append(x)
pari

#list comprehension
pari = [x for x in range(10) if x%2==0]
pari
```

### **3.  Operazioni su array**
Crea un array **a**, 1D (senza digitarlo esplicitamente), e genera due nuovi array **b**, contenente una sottostringa a piacere e **c** che è il reverse dell'array **a** (esempio [1,2,3] diventa [3,2,1]) .  Dividi l'array **a** per l'array **c**. Fai la stessa cosa per una lista.

```python
#creo l'array a da 0 a 4 inclusi
import numpy as np
a = np.arange(1,6)
#creo b con lo slicing
b = a[1:4]
#creo c facendo il reverse di a
c = a[::-1]
#divido l'array a per c
divis_array = a/c
#stampo tutti i risultati
print(a)
print(b)
print(c)
print(divis_array)
#LISTA
#creo la lista a da 0 a 4 inclusi
lista_a = list(range(1,6))
#creo b con lo slicing
lista_b = lista_a[1:4]
#creo c facendo il reverse di a
lista_c = lista_a[::-1]
print(lista_a)
print(lista_b)
print(lista_c)
#divisione
risultato = [lista_a[i]/lista_c[i] for i in range(len(lista_a))]
print(risultato)

```
